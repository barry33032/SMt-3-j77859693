1.Write a program to compute and plot V as a function of P at constant T for the van der Waals gas.

For van der Waals gas,

$$ P=-(\frac{\partial F}{\partial V})_{S,N} = \frac{-a N^2}{V^2}+\frac{k_B N T}{ V - b N} $$

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
T = 0.1
kB=1
a=10**-6
b=10**(-5)
N=10**5

#P = np.arange(0.01, 2.0, 0.01)
V= np.linspace(10, 20, 200)
P = -a*(N**2) /(V**2) +kB*N*T/(V-b*N)

# Note that using plt.subplots below is equivalent to using
# fig = plt.figure() and then ax = fig.add_subplot(111)
fig, ax = plt.subplots()
ax.plot(P,V, '-*b', label='T=0.1 ,a=1 ,b=10^(-5),N=10^5')

ax.set(xlabel='P (some unit)', ylabel='V (some unit)',
       title='PV curve for van der Waals gas')
ax.legend()
ax.grid()

fig.savefig("test.png")
plt.show()
```
